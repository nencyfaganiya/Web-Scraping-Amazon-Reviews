```python
inurl = input("Enter url : ")
cust_name = []
ratings = []
rev_title = []
rev_date = []
rev_content = []
error = []
for i in range(1,51):
  url = inurl+str(i)
  page = requests.get(url)
  if str(page) == "<Response [200]>":
    soup = BeautifulSoup(page.content,'html.parser')
    names= soup.select('span.a-profile-name')[2:]
    stars = soup.select('i.review-rating span.a-icon-alt')[2:]
    title = soup.select('a.review-title-content span')
    date = soup.select('span.review-date')[2:]
    reviews = soup.select('span.review-text-content span')
    for j in range(len(names)):
      cust_name.append(names[j].get_text())
      ratings.append(stars[j].get_text().replace(" out of 5 stars",""))
      rev_title.append(title[j].get_text())
      rev_date.append(date[j].get_text().replace("Reviewed in India on ",""))
      rev_content.append(reviews[j].get_text().strip("\n "))
  else:
    error.append(i)
```
